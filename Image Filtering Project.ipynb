```python
number = int(input("Enter a number: "))
sum_of_digits = 0

while number > 0:
    digit = number % 10
    sum_of_digits += digit
    number //= 10

print("Sum of digits:", sum_of_digits)
```

```python
def fibonacci(n):
    if(n <= 1):
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
n = int(input("Enter number of terms:"))
print("Fibonacci sequence:")
for i in range(n):
    print(fibonacci(i))
```

```python
def fib(n):
    if(n <= 1):
        return n
    else:
        return(fib(n-1) + fib(n-2))
n = int(input('enter a fib series'))
for i in range(n):
    print(fib(i))
```

```python
# how to read file and save the data into csv file format
import csv

# Collect employee information
employee_name = input("Enter employee name: ")
employee_contact = input("Enter employee contact information: ")
employee_job_title = input("Enter employee job title: ")
employee_salary = input("Enter employee salary: ")
employee_tax_info = input("Enter employee tax information: ")

# Write employee information to CSV file
with open('employee_data.csv', mode='a') as employee_file:
    employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    employee_writer.writerow([employee_name, employee_contact, employee_job_title, employee_salary, employee_tax_info])
```

```python
def is_armstrong_number(num):
    # Convert the number to a string to count the digits
    num_str = str(num)
    # Calculate the power to raise each digit to
    power = len(num_str)
    # Calculate the sum of the powered digits
    sum_of_digits = sum([int(digit)**power 
                         for digit in num_str])
    # Return whether the sum of powered digits is equal to the original number
    return sum_of_digits == num
is_armstrong_number(153)
```

```python
# find the element from the list and add them
so = [2,'s',66,'gggg',45,'gfdgdfgdf',88,'sohan',44]

total = 0
for element in so:
    if isinstance(element, int):
        total = total + element
    elif isinstance(element, str) and element.isnumeric():
        total = total + int(element)
    
print(total)
```

```python
# sum of all the even number from list
def sum_even_numbers(numbers):
    total = 0
    for number in numbers:
        if number % 2 ==0:
            total = total + number
    return total

numbers = [1, 2, 3, 4, 5, 6]
sum_even_numbers(numbers)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans

image = mpimg.imread("s.jpeg")
w, h, d = image.shape
pixels = np.reshape(image, (w * h, d))

n_colors = 10 # change this to get more or gewer colors
model = KMeans(n_clusters=n_colors, random_state=42,n_init='auto').fit(pixels)
palette = np.uint8(model.cluster_centers_)

plt.imshow([palette])
plt.show()
```
